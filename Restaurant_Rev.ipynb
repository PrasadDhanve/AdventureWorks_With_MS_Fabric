```python
# Code generated by Data Wrangler for pandas DataFrame

def clean_data(df):
    # Select column: 'Cuisine_Type'
    df = df.loc[:, ['Cuisine_Type']]
    # Drop duplicate rows across all columns
    df = df.drop_duplicates()
    return df

df_clean = clean_data(df.copy())
df_clean.head()
```
